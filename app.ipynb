```python
import os
from pathlib import Path
import logging
```

# logging string

```python
logging.basicConfig(level=logging.INFO, format='[%(acstime)s]:%(message)s')
```

```python
project_name='Kidney_disease_classifier'
```

```python
list_of_files=['.github/workflow/.gitkeep',
            f'src/{project_name}/__init__.ipynb',
            f'src/{project_name}/components/init.ipynb',
            f'src/{project_name}/utils/__init__.ipynb',
            f'src/{project_name}/config/__init__.ipynb',
            f'src/{project_name}/config/confguration.ipynb',
            f'src/{project_name}/pipeline/__init__.ipynb',
            f'src/{project_name}/entity/__init__.ipynb',
            f'src/{project_name}/constants/__init__.ipynb',
            'cofig/config.yaml',
            'dvc.yaml',
            'params.yaml',
            'requirements.txt',
            'setup.ipynb',
            'research/trials.ipynb',
            'templates/index.html']
```

```python
for filepath in list_of_files:
    filepath=Path(filepath)
    filedir, filename=os.path.split(filepath)
    
    if filedir != '':
        os.makedirs(filedir, exist_ok=True)
        logging.info(f'Creating directory; {filedir} for the file: {filename}')
    if not os.path.exists(filepath) or (os.path.getsize(filepath)==0):
        with open(filepath, 'w') as f:
            pass
    else:
        logging.info(f'{filename} already exists.')
```
